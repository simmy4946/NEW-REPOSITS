Is there any noticeable price difference between Walmart and Amazon for the same products?

```python
df1=df[['WALMART_PRICE_CURRENT','AMAZON_PRICE_CURRENT']]
df1
```

```python
df['DIFFERENCE_OF_PRICES']=df['WALMART_PRICE_CURRENT']-df['AMAZON_PRICE_CURRENT']
```

```python
df2=df['DIFFERENCE_OF_PRICES']
df2.reset_index()
```

```python
final_df = pd.concat([df1, df2], axis=1) # row wise concate
final_df
#merged_df = pd.merge(walmart_df, amazon_df, on='Product')
```

```python
mean_difference=final_df['DIFFERENCE_OF_PRICES'].mean()
mean_difference
```

 If the value is positive, it indicates that Walmart prices are higher on average compared to Amazon prices for the same products. Conversely, if the value is negative, it suggests that Walmart prices are lower on average compared to Amazon prices.

```python
median_difference=final_df['DIFFERENCE_OF_PRICES'].median()
median_difference
```

If the median price difference is indeed zero, it may still be valuable to investigate further to understand the underlying factors contributing to this equilibrium in pricing between Walmart and Amazon for the products in the dataset.

```python
std_difference=final_df['DIFFERENCE_OF_PRICES'].std()
std_difference
```

A higher standard deviation indicates greater variability or dispersion in the price differences between Walmart and Amazon for the same products. It suggests that the price differences are more spread out from the mean difference.
